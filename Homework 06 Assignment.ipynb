#### Question 1 - Read in the data set and print the % empty values for each column. Are there any columns with missing values?

```python
df = pd.read_csv('C:\\Users\\birju\\Documents\\CS 301\\groundhog.csv')
df
```

```python
def  get_percent_nans(df, column_name):
    colSize = df.shape[0]
    empty = df.loc[df[column_name].isna()]
    emptySize = empty.shape[0]
    return (emptySize/colSize) * 100
```

```python
for col in df.columns:
    print('The \"{}\" column has {:.2f}% empty values'.format(col, get_percent_nans(df, col)))
```
